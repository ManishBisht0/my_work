# Create a New table in Database

```python
import mysql.connector
import pandas as pd

def create_connection():
    connection = mysql.connector.connect(
        host="databa.amazonaws.com",
        user="admin",
        password="jipoihdoad",
        database="databse",
        port=3307
    )
    return connection

try:
    # Create a connection
    connection = create_connection()

    # Create a cursor
    cursor = connection.cursor()

    # Example SQL query to create a new table
    create_table_query = """
    CREATE TABLE IF NOT EXISTS plants (
        id INT AUTO_INCREMENT PRIMARY KEY,
        Plant_ID INT NOT NULL,
        Plant_Name VARCHAR(255) NOT NULL,
        Image_Url VARCHAR(1000),
        Scientific_Name VARCHAR(255),
        Category VARCHAR(255),
        Color VARCHAR(255),
        Standard_Pot_Size VARCHAR(50),
        Plant_Height_Above_Pot VARCHAR(50),
        Care_and_maintenance VARCHAR(255),
        Learner_Friendly VARCHAR(50),
        Sunlight_Needs VARCHAR(255),
        Watering_Schedule VARCHAR(255),
        Medication_Schedule VARCHAR(255),
        Environment VARCHAR(255),
        Growth_Rate VARCHAR(50),
        Flowering_Season VARCHAR(255),
        Uses VARCHAR(255)
    )
    """

    # Execute the query to create the new table
    cursor.execute(create_table_query)

    # Read data from CSV file into a pandas DataFrame
    csv_file_path = r"C:\Users\DELL\Downloads\wedecor\wedecor8_data.csv"  
    df = pd.read_csv(csv_file_path)

    # Insert data into the table row by row
    for index, row in df.iterrows():
        insert_data_query = f"""
        INSERT INTO plants (
            Plant_ID, Plant_Name, Image_Url, Scientific_Name, Category, Color,
            Standard_Pot_Size, Plant_Height_Above_Pot, Care_and_maintenance,
            Learner_Friendly, Sunlight_Needs, Watering_Schedule, Medication_Schedule,
            Environment, Growth_Rate, Flowering_Season, Uses
        ) VALUES (
            {row['Plant_ID']}, '{row['Plant_Name']}',
            '{row['Image_Url']}', '{row['Scientific_Name']}', '{row['Category']}',
            '{row['Color']}', '{row['Standard_Pot_Size']}', '{row['Plant_Height_Above_Pot']}',
            '{row['Care_and_maintenance']}', '{row['Learner_Friendly']}', '{row['Sunlight_Needs']}',
            '{row['Watering_Schedule']}', '{row['Medication_Schedule']}', '{row['Environment']}',
            '{row['Growth_Rate']}', '{row['Flowering_Season']}', '{row['Uses']}'
        )
        """

        # Execute the query to insert data
        cursor.execute(insert_data_query)

    # Commit the changes
    connection.commit()

    print("Table 'plants' created and data from CSV file inserted successfully.")

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    # Close the cursor and the connection in the 'finally' block to ensure it happens even if an exception occurs
    if cursor:
        cursor.close()
    if connection:
        connection.close()
```
